```python
import sys


def choose_lesson(lesson_list, choosen_lessons, count2, count1=0):
    for i in range(count1, count2):
        choose = input(f"Lesson {i + 1}: ").upper()
        if choose == "Q" or choose == "q":
            break
        if choose in choosen_lessons:
            print("This lesson choosed already, please choose another lesson! If you don't want to choose, press Q.")
            choose_lesson(lesson_list, choosen_lessons, count2, count1=i)
            break
        if choose not in lessons_list:
            print("This lesson is not avaible, please choose an avaible lesson! If you don't want to choose, press Q.")
            choose_lesson(lesson_list, choosen_lessons, count2, count1=i)
            break
        else:
            choosen_lessons.append(choose)
    if len(choosen_lessons) < 3:
        sys.exit("You failed in class.")
    return choosen_lessons


def display_grade(final_grade):
    if (final_grade < 30):
        return "FF You Failed your lesson. Feelsbadman :/ )"
    elif (30 <= final_grade < 50):
        return "DD"
    elif (50 <= final_grade < 70):
        return "CC"
    elif (70 <= final_grade < 90):
        return "BB"
    elif (final_grade >= 90):
        return "AA"
    else:
        return "Wrong Value Please try Again"


student_names = ["MERT SARIKAYA"]
check = False
for i in range(3):
    name = input("Please enter your name: ").upper()
    if name in student_names:
        print(f"Welcome {name}")
        check = True
        break
    print("You entered wrong, please try again")
if not check:
    sys.exit("Please try again later")

lessons_list = ["MATHEMATICS", "STATISTICS", "PHYSICS", "CHEMISTRY", "BIOLOGY"]
choosen_lessons = []

try:
    count = int(input("How many lessons want you choose (minimum 3, maximum 5): "))
except ValueError:
    sys.exit("Lesson count must be integer!!!")
if count < 3 or int(count) > 5:
    sys.exit("You failed in class!")
print("Please choose your lessons: ")
for lesson in lessons_list:
    print("-", lesson)

choosen_lessons = choose_lesson(lessons_list, choosen_lessons, int(count))
lesson_grades = {}

grade_lesson = input("Which lesson want to enter grades: ")
if grade_lesson not in choosen_lessons:
    sys.exit("You not take this lesson!")

try:
    lesson_grades['midterm'] = int(input("Please enter your midterm grade: "))
    lesson_grades['final'] = int(input("Please enter your final grade: :"))
    lesson_grades['project'] = int(input("Please enter your project grade: "))
except ValueError:
    sys.exit("Grades must be integer!!!")

final_grade = lesson_grades['midterm'] * 0.3 + lesson_grades['final'] * 0.5 + lesson_grades['project'] * 0.21

print(f"Your final grade: {final_grade}, your final score: {display_grade(final_grade)}")
```
